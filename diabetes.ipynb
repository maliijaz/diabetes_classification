```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
for i in range(len(df.columns)):
      label = df.columns[i]
      plt.hist(df[df['Outcome']==0][label], label='No Diabetes', color='blue', alpha=0.6, density=True, bins=15)
      plt.hist(df[df['Outcome']==1][label], label='Diabetes', color='red', alpha=0.6, density=True, bins=15)
      plt.title(label)
      plt.ylabel('Number of Patients')
      plt.xlabel(label)
      plt.legend()
      plt.show()
```

```python
x = df.drop('Outcome', axis=1).values
y = df['Outcome'].values
```

```python
y
```

```python
scalar = StandardScaler()
x = scalar.fit_transform(x)
```

```python
over = RandomOverSampler()
x, y = over.fit_resample(x, y)
```

```python
y
```

```python
x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.6, random_state=0)
x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.5, random_state=0)
```

```python
model = tf.keras.models.Sequential([
      tf.keras.layers.Dense(16, activation='relu'),
      tf.keras.layers.Dense(16, activation='relu'),
      tf.keras.layers.Dense(1, activation='sigmoid')
      ])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0002), loss=tf.losses.binary_crossentropy, metrics=['accuracy'])
```

```python
model.evaluate(x_val, y_val)
```

```python
model.fit(x_train, y_train, epochs=30, batch_size=16, validation_data=(x_val, y_val))
```

```python
model.evaluate(x_test, y_test)
```
